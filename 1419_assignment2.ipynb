```python
def read_files(file_loc):
  dataset = []
  
  with open(file_loc, 'r', encoding='cp1258') as test_file:
    for line in test_file:
      dataset.append(line)
    
  return dataset
```

```python
 dataset =pd.read_csv('data.tsv' ,sep='\t')
```

```python
arr=np.array(dataset)
# print(arr)
# print(arr.shape)
inArr=arr
outArr=arr
```

```python
inputSet=np.delete(inArr,8,1)
#print(array.shape)
#print(array)
outputSet=outArr[: ,[8]]
#print(outputset)
```

```python
trainData, testData, trainOut, testOut = train_test_split(inputSet, outputSet, test_size=0.2,random_state=12)
```

```python
model = keras.Sequential(
    [
      #  keras.layers.Flatten()
      keras.layers.Dense(units=1000, input_dim=8),
      keras.layers.Dense(units=500),
     keras.layers.Dense(units=250),
      keras.layers.Dense(units=50),
      keras.layers.Dense(units=10),
      keras.layers.Dense(units=3),
      keras.layers.Dense(units=2),
      keras.layers.Dense(units=1, activation=tf.keras.activations.sigmoid, name='output_layer')
    ]
)

# model.add(keras.layers.Dense(units=1, use_bias=False, activation='relu', name='output_layer'))
print(model.summary())
```

```python
dataset.head(768)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
# Train the NN
model.fit(trainData, trainOut, validation_data=(testData, testOut), epochs=500, verbose=1)
```

```python
print(testOut)
print(model.predict(testData))
```

```python
_, accuracy = model.evaluate(testData, testOut)
print('Accuracy: %.2f' % (accuracy*100))
```
